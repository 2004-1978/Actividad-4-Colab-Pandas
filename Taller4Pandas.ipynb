```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cracion de Dataframe
datos = {
 "Mes": [
 "Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
 "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"
 ],
 "Portatiles": [20, 18, 25, 30, 28, 35, 40, 38, 36, 34, 32, 30],
 "Monitores": [15, 16, 14, 18, 20, 22, 24, 23, 21, 19, 18, 17],
 "Teclados": [40, 42, 38, 45, 47, 50, 55, 53, 52, 48, 46, 44]
}

df = pd.DataFrame(datos)

# Guarda el DataFrame 'df' en un archivo CSV llamado 'ventas_tecnologia.csv'
# index=False evita que el índice de filas de pandas (0, 1, 2, ...) se guarde como una columna en el archivo.
df.to_csv('ventas_tecnologia.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Importar Datos de las Ventas
df = pd.read_csv('ventas_tecnologia.csv')

#Estadisticas De Portatiles
ventas_portatiles = df["Portatiles"].values # convierte la Serie a array de numpy
media_port = np.mean(ventas_portatiles)
mediana_port = np.median(ventas_portatiles)
desv_port = np.std(ventas_portatiles)
print("Estadísticas Portátiles")
print("Media:", media_port)
print("Mediana:", mediana_port)
print("Desviación estándar:", desv_port)

#Estadisticas De Monitores
ventas_monitores = df["Monitores"].values # convierte la Serie a array de numpy
media_port = np.mean(ventas_monitores)
mediana_port = np.median(ventas_monitores)
desv_port = np.std(ventas_monitores)
print("\nEstadísticas Monitores")
print("Media:", media_port)
print("Mediana:", mediana_port)
print("Desviación estándar:", desv_port)

#Estadisticas De Teclados
ventas_teclados = df["Teclados"].values # convierte la Serie a array de numpy
media_port = np.mean(ventas_teclados)
mediana_port = np.median(ventas_teclados)
desv_port = np.std(ventas_teclados)
print("\nEstadísticas Teclados")
print("Media:", media_port)
print("Mediana:", mediana_port)
print("Desviación estándar:", desv_port)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Calcular Ventas Totales por MES
df = pd.read_csv('ventas_tecnologia.csv')

# Calcular las ventas totales por producto
total_portatiles = df['Portatiles'].sum()
total_monitores = df['Monitores'].sum()
total_teclados = df['Teclados'].sum()

# Calcular las ventas totales generales
ventas_totales_generales = total_portatiles + total_monitores + total_teclados

print(f"Ventas totales de Portátiles: {total_portatiles}")
print(f"Ventas totales de Monitores: {total_monitores}")
print(f"Ventas totales de Teclados: {total_teclados}")
print(f"\nLas ventas totales de todos los productos son: {ventas_totales_generales}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Aseguramos que df esté cargado y 'Ventas_Totales' calculado
df = pd.read_csv('ventas_tecnologia.csv')
df['Ventas_Totales'] = df['Portatiles'] + df['Monitores'] + df['Teclados']

# Crear el gráfico de líneas
plt.figure(figsize=(12, 6)) # Tamaño de la figura

plt.plot(df['Mes'], df['Portatiles'], marker='o', label='Portátiles')
plt.plot(df['Mes'], df['Monitores'], marker='o', label='Monitores')
plt.plot(df['Mes'], df['Teclados'], marker='o', label='Teclados')
plt.plot(df['Mes'], df['Ventas_Totales'], marker='o', linestyle='--', color='red', label='Ventas Totales')

# Añadir títulos y etiquetas
plt.title('Ventas Mensuales de Productos de Tecnología')
plt.xlabel('Mes')
plt.ylabel('Cantidad de Ventas')
plt.xticks(rotation=45) # Rotar las etiquetas del eje X para mejor legibilidad
plt.grid(True) # Añadir una cuadrícula
plt.legend() # Mostrar la leyenda para identificar las líneas
plt.tight_layout() # Ajustar el diseño para evitar que las etiquetas se superpongan
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Aseguramos que df esté cargado y 'Ventas_Totales' calculado
df = pd.read_csv('ventas_tecnologia.csv')
df['Ventas_Totales'] = df['Portatiles'] + df['Monitores'] + df['Teclados']

# Crear el gráfico de barras
plt.figure(figsize=(12, 6)) # Tamaño de la figura

plt.bar(df['Mes'], df['Ventas_Totales'], color='skyblue')

# Añadir títulos y etiquetas
plt.title('Ventas Totales Mensuales de Productos de Tecnología')
plt.xlabel('Mes')
plt.ylabel('Cantidad de Ventas Totales')
plt.xticks(rotation=45) # Rotar las etiquetas del eje X para mejor legibilidad
plt.grid(axis='y', linestyle='--', alpha=0.7) # Añadir una cuadrícula horizontal
plt.tight_layout() # Ajustar el diseño para evitar que las etiquetas se superpongan
plt.show()
```
